```python
import json
```

```python
student = {'nazwisko' : 'Kowalski', 'imi' : 'Jan', 'jzyki' : ['angielski', 'niemiecki']}
student
```

```python
s_json = json.dumps(student, ensure_ascii=False, indent = 4, separators = ('', ':'))
print(s_json)
```

```python
s_json = json.dumps(student, ensure_ascii=False, indent = "....", separators = ('', '->'))
print(s_json)
```

```python
s_json = json.dumps(student, ensure_ascii=False, separators = ('', ':'))
print(s_json)
```

```python
dupl_student = json.loads(s_json)
dupl_student
```

```python
with open('test.json','w') as p:
    json.dump(s_json,p)
```

```python
with open('test.json') as p:
    jstudent = json.load(p)
print(jstudent)
```
