```python
class PodKlasa(TakaKlasa):
    def poka_atrybut(self):                           # przykad na polimorfizm
        print(f'Aktualna wartoc = {self.atrybut}')    # zmiana zachowania 
```

```python
pk = PodKlasa()
pk.ustaw_atrybut(22)
pk.poka_atrybut()   # uywamy zmodyfikowanej wersji metody
```

```python
x.poka_atrybut()  #tutaj mamy obiekt klasy TakaKlasa std takie dziaanie metody
```

```python
# podklasa podklasy polimorfizm w peni
```

```python
class PodPodKlasa(PodKlasa):
    def __init__(self,warto):
        self.atrybut = warto
    def __add__(self, skadnik):
        return PodPodKlasa(self.atrybut+skadnik)
    def __str__(self):
        return f'PodPodKlasa:{self.atrybut}'
    def __mul__(self,czynnik):
        return PodPodKlasa(self.atrybut*czynnik)
    
```

```python
a = PodPodKlasa('a')
```

```python
a.atrybut
```

```python
print(a)
```

```python
b = a + 'b'
```

```python
print(b)
```

```python
b4 = b * 4
```

```python
print(b4)
```

```python
c = PodPodKlasa(7)
d = c + 4
e = d * 3
print(e)
```
