```python
class PodKlasa(TakaKlasa):
    def pokaż_atrybut(self):                           # przykład na polimorfizm
        print(f'Aktualna wartośc = {self.atrybut}')    # zmiana zachowania 
```

```python
pk = PodKlasa()
pk.ustaw_atrybut(22)
pk.pokaż_atrybut()   # używamy zmodyfikowanej wersji metody
```

```python
x.pokaż_atrybut()  #tutaj mamy obiekt klasy TakaKlasa stąd takie działanie metody
```

```python
# podklasa podklasy polimorfizm w pełni
```

```python
class PodPodKlasa(PodKlasa):
    def __init__(self,wartość):
        self.atrybut = wartość
    def __add__(self, składnik):
        return PodPodKlasa(self.atrybut+składnik)
    def __str__(self):
        return f'PodPodKlasa:{self.atrybut}'
    def __mul__(self,czynnik):
        return PodPodKlasa(self.atrybut*czynnik)
    
```

```python
a = PodPodKlasa('a')
```

```python
a.atrybut
```

```python
print(a)
```

```python
b = a + 'b'
```

```python
print(b)
```

```python
b4 = b * 4
```

```python
print(b4)
```

```python
c = PodPodKlasa(7)
d = c + 4
e = d * 3
print(e)
```

```python
class rekord: pass  # pusta klasa
```

```python
rekord.nazwa = 'Mirek'      # klasa z atrybutami
rekord.czas = 10.8
```

```python
print(rekord.nazwa)  # działa, choć brak obiektów
```

```python
a = rekord()          # tworzymy puste egzemplarze klasy rekord
b = rekord()
```

```python
a.nazwa, b.nazwa    # ale te puste egzemplarze dają nam dostęp do atrybutów klasy dzięki dziedziczeniu
```

```python
a.nazwa = 'Mietek'                   # do obiektu a dodajemy atrybut nazwa
rekord.nazwa, a.nazwa, b.nazwa       # obiekt y nie ma atrybutu nazwa, więc dziedziczy z klasy rekord
```
