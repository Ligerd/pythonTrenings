# re

```python
import re
```

```python


txt = "Jacek, Wacek i Pankracek"
x = re.findall("acek", txt)
print(x)
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.search("nn", txt)
print(x)
```

```python


txt = "Jacek, Wacek i Pankracek"
x = re.findall("[JWP].", txt)
print(x)
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.search("\s", txt)
print(x.start(), x.end())
print(x.span())
print(x.string)
print(x.group())
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.split("\s", txt)
print(x)
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.split("Pan", txt)
print(x)
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.split("\s", txt, 1)
print(x)
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.sub("\s", "_", txt)
print(x)
```

```python
txt = "Jacek, Wacek i Pankracek"
x = re.sub("\s", "_", txt,2)
print(x)
```
